## Building a Custom Logistic Regression Model with P-Values
**P-value** is used to check whether a feature is statistically significant in the model. A small p-value means the feature’s effect on the target is likely real, while a large p-value means the effect may be random.

```python
from sklearn import linear_model
import scipy.stats as stat

class LogisticRegression_with_PValues:
    def __init__(self, *agrs, **kwargs):
        self.model = linear_model.LogisticRegression(*agrs, **kwargs)

    def fit(self, X, y):
        self.model.fit(X, y)

        # get p-values for the fitted mode
        denom = (2.0 * (1.0 + np.cosh(self.model.decision_function(X))))
        denom = np.tile(denom,(X.shape[1],1)).T
        # fisher Intomation Matrix
        F_ij = np.dot((X /denom).T, X)
        # Inverse Information Matrix
        Cramer_Rao = np.linalg.inv(F_ij)
        signma_estimates = np.sqrt(np.diagonal(Cramer_Rao))
        # z-score for each model coefficient
        z_scores = self.model.coef_[0] / signma_estimates
        # two tailed test for p-values
        p_values = [stat.norm.sf(abs(x)) * 2 for x in z_scores]

        self.coef_ = self.model.coef_
        self.intercept_ = self.model.intercept_
        self.p_values = p_values
```

```python
lg_regession_PValues_model = LogisticRegression_with_PValues()
```

```python
lg_regession_PValues_model.fit(inputs_train, targets_train)
```

```python
summary_table = pd.DataFrame(columns = ['Feature name'], data = feature_names)
summary_table['Coefficients'] = np.transpose(lg_regession_PValues_model.coef_)
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept', lg_regession_PValues_model.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
p_values = lg_regession_PValues_model.p_values
p_values = np.append(np.nan, np.array(p_values))
```

```python
summary_table['p_values'] = p_values
```
