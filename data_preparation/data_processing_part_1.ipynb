### Preprocessing few discrete variables

```python
data.info()
```

We are going to preocess the following discreate variables:
1. **grade** - Main credit risk grade of the borrower.

2. **sub_grade** - More detailed credit grade.

3. **home_ownership** - Borrower’s home ownership status.

4. **verification_status** - Whether borrower’s income was verified.

5. **loan_status** - Current or final loan repayment status.

6. **purpose** - Reason for taking the loan.

7. **addr_state** - Borrower’s U.S. state.

8. **initial_list_status** - Loan listing type: whole or fractional.                                                                                                                                    

Most likely, we are not going to use sub_grade, as it overlaps with grade.

#### Applying ONE-Hot Encoding by pd.get_dummies()

```python
data_dummies = [
    pd.get_dummies(data['grade'], prefix='grade', prefix_sep=':', dtype=int),
    pd.get_dummies(data['sub_grade'], prefix='sub_grade', prefix_sep=':', dtype=int),
    pd.get_dummies(data['home_ownership'], prefix='home_ownership', prefix_sep=':', dtype=int),
    pd.get_dummies(data['verification_status'], prefix='verification_status', prefix_sep=':', dtype=int),
    pd.get_dummies(data['purpose'], prefix='purpose', prefix_sep=':', dtype=int),
    pd.get_dummies(data['addr_state'], prefix='addr_state', prefix_sep=':', dtype=int),
    pd.get_dummies(data['initial_list_status'], prefix='initial_list_status', prefix_sep=':', dtype=int),

]
```

```python
# combining all One_hot encoded data into on dataframe
data_dummies = pd.concat(data_dummies, axis=1)
```

```python
# combining data and data_dummies 
data = pd.concat([data, data_dummies], axis=1)
```

We convert the target column `loan_status` into a numerical binary variable called `good_bad`.

If the loan status represents a bad/default-like borrower, we assign **0**.  
If the loan status represents a good/non-default borrower, we assign **1**.

Examples:

**Charged Off** → 0  
**Default** → 0  
**Does not meet the credit policy. Status:Charged Off** → 0  
**Late (31-120 days)** → 0  

**Fully Paid** → 1  
**Current** → 1  
**In Grace Period** → 1  
**Late (16-30 days)** → 1  
**Does not meet the credit policy. Status:Fully Paid** → 1                                                                                                                                                                            

```python
data['good_bad'] = np.where(
    data['loan_status'].isin([
        'Charged Off', 
        'Default',
        'Does not meet the credit policy. Status:Charged Off',
        'Late (31-120 days)'
        ]
    ),
    0,
    1
)
```

```python
data['good_bad'].value_counts()
```

```python
# Separate the independent variables (features) from the target variable.
# We remove 'good_bad' from X because it is the value we want to predict.
X = data.drop(['good_bad'], axis=1)

# Define the dependent variable (target).
# 'good_bad' contains the class labels: 1 = good borrower, 0 = bad/default-like borrower.
y = data['good_bad']
```
